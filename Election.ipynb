```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('election2024.csv')
```

```python
df.head()
```

```python
df.info(
)
```

```python
df.isnull().values.any()
```

```python
plt.figure(figsize=(8,6))
df['will_vote'].value_counts().plot(kind='bar')
plt.title('Distribution of Voting Intentions')
plt.xlabel('Will Vote')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))  
sns.scatterplot(x='age', y='will_vote', data=df)
plt.title('Age vs Voting Intention')
plt.xlabel('Age')
plt.ylabel('Will Vote')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='education', hue='will_vote', data=df)
plt.title('Education vs Voting Intention')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='television_usage', hue='will_vote', data=df)
plt.title('Television Usage vs Voting Intention')
plt.xlabel('Television Usage')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='internet_usage', y='will_vote', data=df)
plt.title('Internet Usage vs Voting Intention')
plt.xlabel('Internet Usage')
plt.ylabel('Will Vote')
plt.show()
```
